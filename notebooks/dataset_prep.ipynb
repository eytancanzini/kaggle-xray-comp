```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os

%matplotlib inline

TRAIN_DATASETS = '../data/images/train'
TEST_DATASETS = '../data/images/train'

targets = {
    0: "abdomen",
    1: "ankle",
    2: "cervical spine",
    3: "chest",
    4: "clavicles",
    5: "elbow",
    6: "feet",
    7: "finger",
    8: "forearm",
    9: "hand",
    10: "hip",
    11: "knee",
    12: "lower leg",
    13: "lumbar spine",
    14: "others",
    15: "pelvis",
    16: "shoulder",
    17: "sinus",
    18: "skull",
    19: "thigh",
    20: "thoracic spine",
    21: "wrist"
}

!which python3
```

```python
df = pd.read_csv('../data/train_df.csv')
df.shape
```

```python
def imshow(img):
    plt.imshow(img)
    plt.show()

column_names = [
    'SOPInstanceUID',
    'class_id',
    'class_name'
]
img_df = pd.DataFrame(columns=column_names)
    
for filename in os.listdir(TRAIN_DATASETS):
    f = os.path.join(TRAIN_DATASETS, filename)
    if os.path.isfile(f):
        img = cv2.imread(f)
        img_name = filename[:len(filename)-6]
        x = df.loc[df['SOPInstanceUID'] == img_name]
        try:
            id = int(x.iloc[0]['Target'])
            name_id = targets[id]
            temp_1 = {
                'SOPInstanceUID': img_name,
                'class_id': id, 
                'class_name': name_id
            }
            temp = pd.DataFrame(temp_1, index=[0])
            img_df = pd.concat([img_df, temp], ignore_index=True, axis=0)
        except Exception as e:
            print(e)
        
img_df
        
        
```
